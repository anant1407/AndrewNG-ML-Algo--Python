```python
#Load Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load Data
file='ex1data1.txt'
data=np.loadtxt(file,delimiter=',')
X=data[:,:-1]
y=data[:,1] 
y=y[:,np.newaxis] #1D to 2D array
m=X.shape[0] #no. of training examples
n=data.shape[1] #no. of features
```

```python
#Plot data
plt.scatter(X,y, marker='x',c='red')

plt.xlabel('Population of city in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

```python
#add ones to X
x1=np.ones((m,1))
X=np.concatenate((x1,X),axis=1)
theta=np.zeros((n,1))
X.shape
```

```python
#Computation
alpha=0.01
iterations=1500

h=np.matmul(X,theta)
cost= sum((h-y)**2)/(2*m)


for i in range(iterations):
    grad= sum((h-y)*X)
    grad=grad.reshape(n,1)
    theta= theta-((alpha/m)*grad)
    h=np.matmul(X,theta)
    cost= sum((h-y)**2)/(2*m)
    
print(cost)
print(theta)
    
```

```python
#plot curve
plt.scatter(data[:,0],y, marker='x',c='red')
plt.plot(data[:,0],h)
```
