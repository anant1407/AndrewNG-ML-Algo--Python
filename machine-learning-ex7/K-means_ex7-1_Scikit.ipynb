```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
mat=scipy.io.loadmat('ex7data2.mat')
```

```python
X=mat['X']
K=3 #no.of clusters
m=X.shape[0]
n=X.shape[1]
J=0
```

```python
#plot data
plt.style.use('ggplot')
plt.scatter(X[:,0],X[:,1], marker='+')
plt.show()
```

```python
from sklearn.cluster import KMeans
K=3
kmeans=KMeans(n_clusters=K)
y_pred=kmeans.fit_predict(X)
kmeans.labels_  #y predict
kmeans.cluster_centers_  #Centroids of data
```

```python
#plot solution
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],marker='o',c='yellow')
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],marker='+',c='blue')
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],marker='x',c='red')
```
