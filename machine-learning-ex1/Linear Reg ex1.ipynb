```python
#Load Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load Data
file='ex1data1.txt'
data=np.loadtxt(file,delimiter=',')
X=data[:,:-1]
y=data[:,1] 
y=y[:,np.newaxis] #1D to 2D array
m=X.shape[0] #no. of training examples
n=data.shape[1] #no. of features
```

```python
#Plot data
plt.scatter(X,y, marker='x',c='red')

plt.xlabel('Population of city in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

```python
#add ones to X
x1=np.ones((m,1))
X=np.concatenate((x1,X),axis=1)
theta=np.zeros((n,1))
```

```python
#Itereations : 1500
alpha=0.01
iterations=1500

h=np.matmul(X,theta)
cost= sum((h-y)**2)/(2*m)


for i in range(iterations):
    grad= sum((h-y)*X)
    grad=grad.reshape(n,1)
    theta= theta-((alpha/m)*grad)
    h=np.matmul(X,theta)
    cost= sum((h-y)**2)/(2*m)
    
print('Cost: \n',cost)
print('Theta: \n',theta)
```

```python
#plot curve
plt.scatter(data[:,0],y, marker='x',c='red')
plt.plot(data[:,0],h)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression(fit_intercept=True)

model.fit(X,y)

theta0=model.intercept_
theta_rem= model.coef_[0,1:]
theta_scikit=np.concatenate((theta0,theta_rem))
theta_scikit=theta_scikit[:,np.newaxis]
h_scikit=np.matmul(X,theta_scikit)
cost_scikit= sum((h_scikit-y)**2)/(2*m)

print('Cost_Scikit: \n',cost_scikit)
print('Theta_Scikit: \n',theta_scikit)
```

```python
#plotting both solutions
plt.scatter(data[:,0],y, marker='x',c='red')
plt.plot(data[:,0],h)
plt.plot(data[:,0],h_scikit)
```
